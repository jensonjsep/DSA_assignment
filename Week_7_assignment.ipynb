```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### 1. First of all, check whether there is any relationship between the amount spent for advertising through five different companies and profit generated.

```python
corr_ad = data_ad.corr()
corr_ad
```

```python
plt.subplots(figsize=(10,4))
sns.heatmap(corr_ad,fmt=".5g", vmin=-1, vmax=1, annot=True, linewidth= 0.2, cmap='Blues')
```

##### Insight:
Here we a see that all features have high correlation to  each other.

### 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
data1 = data_ad.drop(['Date'],axis=1)
```

```python
y = data1['Profit(in Thousands)']
X = data1.drop(['Profit(in Thousands)'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45, test_size = 0.25)
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
```

### 3. Predict the profit generated by the company.

```python
predictions=model.predict(X_test)
make_predict = [[290,312,309,298,320]]
model.predict(make_predict).item()
```

#### Insight:
Here the prediction of profit for the given amount for advertisement is 5820.06 (in Thousands)
