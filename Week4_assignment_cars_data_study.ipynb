# Case study on Presentation of data

The data set is cars_data.csv

This file contains the sales related data about different cars.

Here we are doing the following operations on ‘cars_data’ dataset:

1. Read the dataset to the python environment.
2. Check for the null values present in the dataset.
3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# 1. Read the dataset to the python environment.

```python
car_data = pd.read_csv('cars_data.csv')
car_data.head()
```

# 2. Check for the null values present in the dataset.

```python
car_data.isna().sum()
```

```python
car_data.info()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.


```python
#creating the chart
plt.figure(figsize=(4,6))
car_plot = sns.countplot(x='Buyer Gender',data = car_data)
car_plot
plt.title('Male vs Female buyers participated',fontweight="bold", pad=20)
plt.ylim(4500, 5500)

# printing the insights
male_car = car_data.loc[car_data['Buyer Gender']== 'Male'].count()
female_car = car_data.loc[car_data['Buyer Gender']== 'Female'].count()
print('Insights:\n')
print('No. of Male buyers =',male_car['Buyer Gender'])
print('No. of Female buyers =',female_car['Buyer Gender'])

# annotating the values of the bars on top of it.
for i in car_plot.patches:
    car_plot.annotate(format(i.get_height(), '.2f'),
                   (i.get_x() + i.get_width() / 2,
                    i.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 30),rotation=15,
                   textcoords='offset points')
```

Insights:

No. of Male buyers = 4948

No. of Female buyers = 5052

 # 4. Find the top 5 cars based on their sales price.

```python
lar_car = car_data.nlargest(5, ['Sale Price'])
lar_car
```

Insight :
The top 5 cars having the highest sales price are : Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES

# 5. Find the least 5 cars based on their Resell price.

```python
sml_car = car_data.nsmallest(5,['Resell Price'])
sml_car
```

Insight : 
5 cars having the lowest Resell price are : Chevrolet 1500, GMC Savana 1500, Jaguar XF, Mistubishi 30000GT and Infiniti G37
