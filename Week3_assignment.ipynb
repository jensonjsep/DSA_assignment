```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Read the dataset to python environment.

```python
iris = pd.read_excel('iris.xls')
iris.head()
```

# Display the columns in the dataset

```python
iris.columns
```

# Calculate the mean of each column of the dataset.

```python
iris1 = pd.DataFrame(iris)
```

```python
print('Mean of SL:',iris1['SL'].mean())
print('Mean of SW:',iris1['SW'].mean())
print('Mean of PL:',iris1['PL'].mean())
print('Mean of PW:',iris1['PW'].mean())
```

# Check for the null values present in the dataset.

```python
iris1.isna().sum()
```

```python
 iris1.info()
```

```python
iris1['Classification'].value_counts()
```

# Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
sns.stripplot(x='Classification',y="SL",data=iris1)
```

```python
sns.violinplot(x='Classification',y="SW",data=iris1)
```

```python
sns.swarmplot(x='Classification',y="PL",data=iris1,s=3)
```

```python
plt.figure(figsize=(6,6))

plt.subplot(2,1,1)
plt.scatter(iris1['Classification'],iris1['PL'],s=4)
plt.title('Classification VS PL representation',fontsize=20)
plt.xlabel('Classification')
plt.ylabel('PL')

plt.subplot(2,1,2)
plt.scatter(iris1['Classification'],iris1['PW'],s=4)
plt.title('Classification VS PW representation',fontsize=20)
plt.xlabel('Classification')
plt.ylabel('PW')

plt.tight_layout()
```
